```python
imigrantes_canada = pd.read_csv("./canadian_immegration_data.csv")
imigrantes_canada.set_index('Country', inplace=True)
anos = list(map(str, range(1980, 2014)))

brasil = imigrantes_canada.loc["Brazil", anos]
argentina = imigrantes_canada.loc["Argentina", anos]

plt.figure(figsize=(12, 6))
plt.plot(anos, brasil.values, label='Brasil')
plt.plot(anos, argentina.values, label='Argentina')

plt.title('ImigraÃ§Ã£o para o CanadÃ¡: Brasil vs. Argentina (1980-2013)')
plt.xlabel('Ano')
plt.ylabel('NÃºmero de Imigrantes')
plt.xticks(anos[::5]) 
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(df_imigrantes_brasil['Ano'], df_imigrantes_brasil['Imigrantes'], label='Brasil')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title ("Imigracao do Brasil para o Canada \n 1980 a 2013")
ax.set_xlabel("Ano")
ax.set_ylabel("Imigrantes")
plt.show()
```

```python


import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm



america_do_sul = imigrantes_canada[imigrantes_canada['Region'] == 'South America']
anos = list(map(str, range(1980, 2014)))


south_america_immigration = america_do_sul[anos].sum(axis=1).sort_values(ascending=False)


colors = cm.get_cmap('viridis', len(south_america_immigration))

plt.figure(figsize=(12, 6))
bars = plt.bar(south_america_immigration.index, south_america_immigration.values, color=colors(range(len(south_america_immigration))))

plt.title('Total de Imigrantes da AmÃ©rica do Sul para o CanadÃ¡ (1980-2013)')
plt.xlabel('PaÃ­ses')
plt.ylabel('NÃºmero de Imigrantes')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout() 

plt.legend(bars, south_america_immigration.index, bbox_to_anchor=(1.05, 1), loc='upper left', title='PaÃ­ses')
plt.show()
```
